```python
# possible ways to identify outliers: IQR upper/lower bounds, z scores
```

```python
X_train.describe()
```

```python
budget_Q3 = X_train["budget"].describe()["75%"]
budget_Q1 = X_train["budget"].describe()["25%"]
budget_median = X_train["budget"].describe()["50%"]
```

```python
budget_IQR = budget_Q3 - budget_Q1
```

```python
runtime_Q3 = X_train["runtime"].describe()["75%"]
runtime_Q1 = X_train["runtime"].describe()["25%"]
runtime_median = X_train["runtime"].describe()["50%"]
```

```python
runtime_IQR = runtime_Q3 - runtime_Q1
```

```python
budget_upper_bound = budget_Q3 + (1.5 * budget_IQR)
budget_lower_bound = budget_Q1 - (1.5 * budget_IQR)
```

```python
runtime_upper_bound = runtime_Q3 + (1.5 * runtime_IQR)
runtime_lower_bound = runtime_Q1 - (1.5 * runtime_IQR)
```

```python
# possible ways to approach outliers: ignore them, remove the rows with outliers, replace the outliers with the median
```

```python
# remove outliers:
# X_train = X_train[(X_train["budget"] > budget_lower_bound) & (X_train["budget"] < budget_upper_bound) & (X_train["runtime"] > runtime_lower_bound) & (X_train["runtime"] < runtime_upper_bound)]
```

```python
# or, replace outliers with the median
```

```python
def replace_outliers(col, upper, lower, median):
    if ((col > upper) | (col < lower)):
        return median
    else:
        return col
```

```python
X_train["budget"] = X_train["budget"].apply(replace_outliers, args=(budget_upper_bound, budget_lower_bound, budget_median))
X_train["runtime"] = X_train["runtime"].apply(replace_outliers, args=(budget_upper_bound, budget_lower_bound, budget_median))
```

## Preprocessing using Pipelines

```python
from sklearn.compose import ColumnTransformer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import StandardScaler

scale_features = ['budget', 'release_date', 'runtime']

preproc = ColumnTransformer([
    # List of transforms/tasks to execute in parallel
    ('text', CountVectorizer(max_features=20, stop_words='english'), 'genres'),
    ('scale', StandardScaler(), scale_features)
])

preproc.fit(X_train)
genre_cols = ['genre_' + word for word in preproc.named_transformers_['text'].get_feature_names_out()]
X_train_preproc = pd.DataFrame(data=preproc.transform(X_train).array())
X_train_preproc.head()
```
